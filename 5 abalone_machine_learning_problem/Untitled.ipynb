```python
df_x=df_abalone.drop(['Rings'],axis=1)
y=df_abalone[['Rings']]
```

```python
for index in df_x.columns:
    if df_x.skew().loc[index]>0.5:
        df_x[index]=np.log1p(df_x[index])
    if df_x.skew().loc[index]<-0.5:
        df_x[index]=np.log1p(df_x[index])
```

```python
df_x.skew()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split
```

```python
def maxr2_score(regr,df_x,y):
    max_r_score=0
    for r_state in range(40,100):
        x_train,x_test,y_train,y_test=train_test_split(df_x,y,test_size=0.20,random_state=r_state)
        regr.fit(x_train,y_train)
        y_pred=regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print("r2 score corresponding to ",r_state," is ",r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("max r2 score corresponding to ",final_r_state," is ",max_r_score)
    return final_r_state
        
```

### Lets use decision tree

```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
dtr=DecisionTreeRegressor()
r_state=maxr2_score(dtr,df_x,np.log1p(y))
```

### lets use random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore")
rfr=RandomForestRegressor()
parameters = {"n_estimators":[10,100,500]}
clf = GridSearchCV(rfr, parameters, cv=5)
clf.fit(df_x, np.log1p(y))
clf.best_params_
```

```python
rfr=RandomForestRegressor(n_estimators=500)
r_state=maxr2_score(rfr,df_x,np.log1p(y))
```
